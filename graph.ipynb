```python
import numpy as np
import matplotlib.pyplot as plt

methods = [
    'Baseline',
    'Low Light',
    'Logarithmic',
    'Homomorphic',
    'Homo + Log'
]

accuracy = [0.9720, 0.7960, 0.8320, 0.8840, 0.8920]
f1_macro = [0.9720, 0.7957, 0.8346, 0.8830, 0.8911]
auc_ovr  = [0.9940, 0.9482, 0.9707, 0.9800, 0.9852]

x = np.arange(len(methods)) 
width = 0.25                 

fig, ax = plt.subplots(figsize=(10, 6))

# Accuracy、F1、AUC
rects1 = ax.bar(x - width, accuracy, width, label='Accuracy')
rects2 = ax.bar(x,        f1_macro, width, label='F1 (macro)')
rects3 = ax.bar(x + width, auc_ovr, width, label='AUC (OVR)')


ax.set_ylabel('Score')
ax.set_title('Performance under Different Enhancements')
ax.set_xticks(x)
ax.set_xticklabels(methods, rotation=15)
ax.set_ylim(0.7, 1.02)  
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.tight_layout()
plt.show()
```
